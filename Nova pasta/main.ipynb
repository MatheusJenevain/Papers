```python
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

# Sample text data
random_phrase = "The bank by the river is steep."

# Tokenizing text into words
tokens = word_tokenize(random_phrase.lower())  # Tokenizing and converting to lowercase

# Function to get context of a word and its meanings
def get_word_meanings(word):
    synsets = wordnet.synsets(word)
    meanings = []
    if synsets:
        for syn in synsets:
            meanings.append(syn.definition())
    return meanings

# Extracting words and their meanings
unique_words = set(tokens)
for word in unique_words:
    meanings = get_word_meanings(word)
    if meanings:
        print(f"Word: '{word}'")
        print(f"Context: '{' '.join(tokens)}'")
        print(f"Meanings:")
        for meaning in meanings:
            print(f"- {meaning}")
        print("-" * 50)
```
